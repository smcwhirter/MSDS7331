```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw_data_url = 'https://raw.githubusercontent.com/andrewmejia600/MSDS7331/master/RAW_DATA/adult.data'
```

```python
raw_census_95_data_names = ['age', 'workclass', 'fnlwght', 'education', 'education_num_yr', 'martial_status', 'occupation', 'relationship', 'race', 'sex', 'cap_gain', 'cap_loss', 'hours_per_week', 'native_country', 'annual_income']
```

```python
raw_census_95_data = pd.read_csv(raw_data_url, names = raw_census_95_data_names)
```

```python
#renaming df to make typing easier
rcd=raw_census_95_data
```

```python
#check for missing values
rcd.isnull().sum(axis=0)
```

```python
#Separating categorical and numeric for EDA purposes
rcd_num=rcd.select_dtypes(include='int64')
rcd_num.head()
```

```python
rcd_cat=rcd.select_dtypes(include='object')
rcd_cat.head()
```

```python
#Examiniation of the various categorical values
rcd_cat.native_country.value_counts()
```

```python
#Distributions of numeric variables
#Code derived from a Toward Data Science blog (link:https://towardsdatascience.com/how-to-perform-exploratory-data-analysis-with-seaborn-97e3413e841d)
rcd.hist(bins=15, figsize=(15,20), layout=(5,2))
```

```python
#pairwise comparison of numeric data
#Code derived from seaborn docs (link:https://seaborn.pydata.org/tutorial/distributions.html)
sns.pairplot(rcd, hue="annual_income")
```

```python
#Counts of the different categorical attributes
#Code derived from a Toward Data Science blog (link:https://towardsdatascience.com/how-to-perform-exploratory-data-analysis-with-seaborn-97e3413e841d)
fig, ax=plt.subplots(3, 3, figsize=(20, 30))
for attribute, subplot in zip(rcd_cat, ax.flatten()):
    sns.countplot(rcd_cat[attribute], ax=subplot)
    for label in subplot.get_xticklabels():
        label.set_rotation(45)
```

```python
#Older broken out by annual_income
rcd["older"]=np.where(rcd["age"]>25,"older","younger")
```

```python
sns.countplot(x="older", hue="annual_income", data=rcd)
```

```python
rcd["o_and_m"]=np.where((rcd["older"]=="older") & (rcd["martial_status"]=="married"),"yes","no")
```

```python
#see if clumping older and married does anything
sns.countplot(x="o_and_m", hue="annual_income", data=rcd)
```

```python
#Separating by capital gains
rcd["bigcap"]=np.where(rcd["cap_gain"]>7000,"yes","no")
```

```python
#bigcap broken out by annual_income
sns.countplot(x="bigcap", hue="annual_income", data=rcd)
```
